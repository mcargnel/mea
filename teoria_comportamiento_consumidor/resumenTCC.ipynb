## Preferencias
Símbolo $\geq$, $>$ ó ~ indican la relación de preferencia. Igual que en matemática se habla de preferencias fuertes y débiles, ahora podemos definir ciertas propiedades

### Necesarias para la existencia de la funcion de utilidad
- Completitud: cuando hay que elegir entre dos opciones tenes que decir si preferis alguna o te da lo mismo. No podes decir que no sabes.
- Reflexividad: una canasta es tan buena como si misma, es decir que si dos canastas son igual te tiene que dar lo mismo cualquiera de las dos.
- Transitividad: Si una canasta A es mejor que una canasta B y B es mejor que una canasta C. Entonces A es mejor que C.

### Cuanto elige el individuo
- Saciedad y no saciedad local: hay un punto que deja satisfecho al individuo. No saciedad local indica que dada una canasta, siempre existe otra lo suficientemente cerca que es mejor.



- Monotonicidad: Una canasta solo es mejor si tiene más de todos los bienes. Fuerte se considera cuando me das más de todo, sino es solo monotonicidad.
